```python
req_cols = ['District code','District name','Population','Male','Female','Literate','Households_with_Internet','Housholds_with_Electric_Lighting']
props = props[req_cols]
props.head()
```

```python
df = dist.merge(props,left_on='District',right_on='District name').drop(columns=['District name'])
```

```python
df.head()
```

```python
df['sex_ratio'] = round(df['Female']/df['Male']*100,2)
df['Literatcy_rate'] = round(df['Literate']/df['Population']*100,2)
df.head()
```

```python
df.to_csv('india.csv',index=False)
```
